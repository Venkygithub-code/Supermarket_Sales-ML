```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
df = pd.read_csv("super.csv")
df.head(10)
```

```python
df.tail(5)
```

```python
df.info()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Branch', data=df, palette='Set1')
plt.title('Number of Sales by Branch')
plt.xlabel('Branch')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
df['Customer type'].value_counts().plot(kind='pie', autopct='%1.2f%%', colors=['skyblue', 'lightgreen'])
plt.title('Distribution of Customer Types')
plt.ylabel('')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df, palette='pastel')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['Unit price'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Product Prices')
plt.xlabel('Unit Price ($)')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='Rating', data=df, color='lightblue')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.pairplot(df[['Unit price', 'Quantity', 'Tax 5%', 'Total', 'cogs', 'gross income', 'Rating']], diag_kind='kde')
plt.suptitle('Pairplot of Numeric Variables', y=1.02)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='Product line', hue='Gender', data=df, palette='Set2')
plt.title('Product Line Distribution by Gender')
plt.xlabel('Product Line')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df_time = df.groupby('Date')['Total'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.lineplot(x='Date', y='Total', data=df_time, color='green')
plt.title('Total Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales ($)')
plt.xticks(rotation=45)
plt.show()
```

```python
numeric_cols = df.select_dtypes(include=['float64', 'int64'])

# Calculate correlation matrix
corr = numeric_cols.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```
